```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv('train.csv.zip', compression='zip')
```

```python
train.shape
```

```python
train.head()
```

```python
def count_missing(df):
    print(df.name, df.isnull().sum())

for column in train:
    count_missing(train[column])
```

```python
target = train['Category']
train.drop(['Category'], axis=1, inplace=True)
train.drop(['Descript'], axis=1, inplace=True)
train.drop(['Resolution'], axis=1, inplace=True)
```

```python
target.describe()
```

```python
target.value_counts()
```

```python
train['DayOfWeek'].value_counts()
```

```python
pd.crosstab(train['DayOfWeek'], target)
```

```python
pd.crosstab(train['PdDistrict'], target)
```

```python
X = pd.get_dummies(train['DayOfWeek'], columns=train.columns)
```

```python
X = X.join(pd.get_dummies(train['PdDistrict'], columns=train.columns))
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(target)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=123, test_size=0.3)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import log_loss
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
y_pred = np.array(model.predict_proba(X_test))
print("Log loss "+str(log_loss(y_test, y_pred)))
```
